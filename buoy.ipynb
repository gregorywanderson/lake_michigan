# NDBC Data

This notebook introduces National Data Buoy Center (NDBC) data with a focus on buoys in Southern Lake Michigan.

>[The mission of NOAA's NDBC is to collect and disseminate in-situ, real-time, quality-controlled observations in the marine environment to ensure the Nation's maritime safety, and to understand and predict the atmosphere, ocean, waves, ice, and climate.](https://www.ndbc.noaa.gov/faq/ndbc.shtml) 

The NDBC provides both station metadata and station measurements. NDBC data, guides and documentation can be found on the NDBC web site. 

- [National Data Buoy Center](https://www.ndbc.noaa.gov/)
- [NDBC Web Data Guide](https://www.ndbc.noaa.gov/docs/ndbc_web_data_guide.pdf)
- [Measurement Description and Units](https://www.ndbc.noaa.gov/faq/measdes.shtml)
- [Station List](https://www.ndbc.noaa.gov/to_station.shtml)
- [Realtime data from the last 45 days](https://www.ndbc.noaa.gov/data/realtime2/)
- [Historical Data](https://www.ndbc.noaa.gov/historical_data.shtml)

The Standard Meteorological data includes:

- *WDIR* &nbsp; Wind direction
- *WSPD* &nbsp; Wind speed (m/s) averaged over an eight-minute period for buoys and a two-minute period for land stations. 
- *GWS* &nbsp; Peak 5 or 8 second wind gusts speed (m/s).
- *WVHT* &nbsp; Significant wave height (m)
- *DPD* &nbsp; Dominant wave period (s)
- *APD* &nbsp; Average wave period (s)
- *PRES* &nbsp; Sea level pressure (hPa)
- *ATMP* &nbsp; Air temperature (Celsius)
- *WTMP* &nbsp; Water temperature (Celsius)
- *DEWP* &nbsp; Dewpoint

Solar Radiation data may include:

- *SRAD1* &nbsp; Shortwave Radiation, 
[LI-COR LI-200 pyronometer sensor](https://www.licor.com/env/products/light/pyranometer.html), $0.4$-$1.1~\mu{\rm m}$ ``solar'' (${\rm W}/{\rm m}^2$)
- *SWRAD* &nbsp; Shortwave Radiation, 
[Eppley Precision Spectral Pyranometer (PSP)](https://www.sikuliaq.alaska.edu/ops/pdf/Eppley_PSP_and_PIR_datasheets.pdf), $0.285$-$2.8~\mu{\rm m}$ ``solar'' (${\rm W}/{\rm m}^2$)
- *LWRAD* &nbsp; Longwave Radiation, 
[Eppley Precision Infrared Radiometer (PIR)](http://www.eppleylab.com/wp-content/uploads/pdf/EPPLEY160919-PIR.pdf), IR from the Earth and atmosphere, $4$-$50~\mu{\rm m}$  (${\rm W}/{\rm m}^2$)

See the [Measurement Description and Units](https://www.ndbc.noaa.gov/faq/measdes.shtml) link for more information.


The NDBC web interface (https://www.ndbc.noaa.gov/) which is linked in the image below, is a convenient way to identify stations.

## Wave height

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from matplotlib.colors import LinearSegmentedColormap

def plot_waveheight_vs_period(buoydata: pd.DataFrame, 
                              station_id: str = None, 
                              title: str = "Wave Height and Period"):
    """
    Expected columns in buoydata:
      - 'DPD' (dominant wave period, seconds)
      - 'WVHT'  (meters)
      - 'datetime'     (datetime-like; used only to extract the year for the annotation)
    """

    # --- Validate/prepare data ---
    required = ["DPD", "WVHT"]
    for c in required:
        if c not in buoydata.columns:
            raise ValueError(f"buoydata must contain column '{c}'")

    x = pd.to_numeric(buoydata["DPD"], errors="coerce").dropna()
    y = pd.to_numeric(buoydata["WVHT"], errors="coerce").dropna()

    # Keep only rows where both x and y are available
    valid_idx = x.index.intersection(y.index)
    x = x.loc[valid_idx].to_numpy()
    y = y.loc[valid_idx].to_numpy()

    # Extract year for annotation (mimic R's buoydata$datetime[2])
    year_text = ""
    if "datetime" in buoydata.columns:
        try:
            dt = pd.to_datetime(buoydata.loc[valid_idx, "datetime"])
            # Use the second valid entry if possible, else first
            if len(dt) >= 2:
                year_text = str(pd.to_datetime(dt.iloc[1]).year)
            elif len(dt) >= 1:
                year_text = str(pd.to_datetime(dt.iloc[0]).year)
        except Exception:
            pass

    # --- Build a grid for KDE (similar to stat_density2d with n=200) ---
    # Use robust bounds to avoid extreme outliers exploding the grid
    x_min, x_max = np.nanpercentile(x, [1, 99])
    y_min, y_max = np.nanpercentile(y, [1, 99])
    
    x_min = x.min()
    y_min = y.min()
    x_max = x.max()
    y_max = y.max()

    xi = np.linspace(x_min, x_max, 200)
    yi = np.linspace(y_min, y_max, 200)
    xx, yy = np.meshgrid(xi, yi)

    # KDE over the grid
    kde = gaussian_kde(np.vstack([x, y]))
    zz = kde(np.vstack([xx.ravel(), yy.ravel()])).reshape(xx.shape)

    # Match ggplot's fill=..density..^0.25
    zz = np.power(zz, 0.25)

    # --- Colormap roughly from white to aquamarine-like teal ---
    cmap = LinearSegmentedColormap.from_list(
        "white_to_aqua",
        [(1, 1, 1), (0.02, 0.4, 0.4)]  # white -> aquamarine-ish
    )

    # --- Plot ---
    fig, ax = plt.subplots(figsize=(7.5, 5.5), dpi=140)

    # Density tiles (use pcolormesh for a "tile" feel)
    pcm = ax.pcolormesh(xi, yi, zz, shading="auto", cmap=cmap)

    # Scatter overlay, similar to geom_point(alpha=0.05, shape=20, size=1)
    ax.scatter(x, y, s=6, alpha=0.05, marker="o", linewidths=0)

    # Labels and title
    ax.set_title(title)
    ax.set_xlabel("Dominant Wave Period (s)")
    ax.set_ylabel("Wave Height (meters)")

    # Annotations roughly akin to your fixed positions (adjust dynamically near top-left)
    xr = x_max - x_min
    yr = y_max - y_min
    x_left = x_min + 0.02 * xr
    y_top  = y_max - 0.06 * yr

    ax.text(x_left, y_top, "Buoy:", fontsize=9, va="center", ha="left")
    if station_id:
        ax.text(x_left + 0.10 * xr, y_top, station_id, fontsize=9, va="center", ha="left")
    if year_text:
        ax.text(x_left + 0.65 * xr, y_top, year_text, fontsize=9, va="center", ha="left")

    # Clean theme similar to theme_classic()
    for spine in ["top", "right"]:
        ax.spines[spine].set_visible(False)

    # Optional colorbar (often useful)
    cbar = fig.colorbar(pcm, ax=ax, shrink=0.9, pad=0.02)
    cbar.set_label("Density^0.25")

    plt.tight_layout()
    plt.show()
```

```python
# pip install pandas matplotlib scipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

def plot_wspd_histogram(buoydata: pd.DataFrame, station: str = "", show_weibull: bool = True):
    """
    Assumes buoydata['WSPD'] is wind speed in knots (to match the R x-label).
    """

    if "WSPD" not in buoydata.columns:
        raise ValueError("buoydata must contain a 'WSPD' column")

    # Clean numeric and drop NaNs
    wspd = pd.to_numeric(buoydata["WSPD"], errors="coerce").dropna().to_numpy()

    # Histogram settings to match R code
    n_bins = 30
    x_min, x_max = 0.0, 20.0
    binwidth = (x_max - x_min) / n_bins

    # Colors close to R's 'darkmagenta' and 'purple4'
    fill_color = "#8B008B"    # darkmagenta
    edge_color = "#551A8B"    # purple4

    fig, ax = plt.subplots(figsize=(7.5, 5))
    counts, bins, patches = ax.hist(
        wspd, bins=n_bins, range=(x_min, x_max),
        color=fill_color, edgecolor=edge_color, linewidth=0.5,
        density=False  # freq=TRUE in R → counts, not density
    )

    # Titles / labels
    ax.set_title("Distribution of Wind Speeds", pad=20)
    if station:
        # mimic R's "sub" by adding a smaller subtitle-like text under the title
        ax.text(0.5, 1.01, station, transform=ax.transAxes, ha="center", va="bottom", fontsize=10)
    ax.set_xlabel("Wind Speed (m/s)")
    ax.set_xlim(x_min, x_max)
    ax.set_ylabel("Count")

    # Optional: fit and overlay a Weibull distribution (on counts scale)
    if show_weibull and np.any((wspd >= x_min) & (wspd <= x_max)):
        data = wspd[(wspd >= x_min) & (wspd <= x_max)]
        # Constrain location to 0 to match wind speed being non-negative
        k, loc, scale = weibull_min.fit(data, floc=0)
        x = np.linspace(x_min, x_max, 400)
        pdf = weibull_min.pdf(x, k, loc=0, scale=scale)
        # Convert PDF to histogram counts: n * binwidth * pdf(x)
        y = pdf * len(data) * binwidth
        ax.plot(x, y, lw=2, label=f"Weibull fit (k={k:.2f}, scale={scale:.2f})")
        ax.legend()

    # Clean look similar to theme_classic()
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    plt.tight_layout()
    plt.show()

# Example:
plot_wspd_histogram(buoydata, station="46007", show_weibull=True)
```

## Python APIs for the NDBC

Thare are several python packages that serve as APIs for the NBDC including:

* [Siphon](https://unidata.github.io/siphon/latest/examples/ndbc/buoy_type_request.html) a collection of Python utilities for downloading data from remote data services. This 
[unidata](https://www.unidata.ucar.edu/) packages has good search functions. Data retrieval currently seems to be implemented only for realtime data. Unidata is a [UCAR](https://www.ucar.edu/who-we-are) community program.
See also 

  - Nested

  - [Using Siphon to query the NetCDF Subset Service](https://notebook.community/ocefpaf/notebooks_demos/notebooks/staged_noteooks/SIPHON_NCSS_HFR_DataAccess)


* [ndbc-api](https://github.com/cdjellen/ndbc-api), this Github also contains [Example Notebook](https://github.com/CDJellen/ndbc-api/blob/main/notebooks/overview.ipynb). ndbc-api also has a [PyPi Repository](https://pypi.org/project/ndbc-api/). The station search methods in ndbc_api return very incomplete results. In particular, as of February 2024, the stations() method currently returns 146  out of 1323 stations, and nearest_station() method seems restricted to that subset.  The ndbc_api methods which access station data and metadata seem to work for the complete set of buoy's provided you have the five digit station identifier. Those methods are.

    - station() which returns station metadata
    - available_realtime() which can return data from the last 45 days
    - available_historical() which returns older data.

* [NDBC](https://pypi.org/project/NDBC/) A pacakge to simplify to retrieval and parsing of NOAA NDBC data.

* [seebuoy](https://www/seebuoy.com) Easily access real time and historical data from the National Data Buoy Center.
     - [GitHub](https://github.com/nickc1/seebuoy)
